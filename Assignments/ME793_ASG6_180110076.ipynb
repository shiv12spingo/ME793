###Q1.

Make sure that you are using the data that you created for Assignment 4.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/ME793/element_data.csv')
display(df)
```

Remove all the columns except elements and Youngs modulus. Sort the table according to the increasing order of Youngs modulus.

```python
df2 = df[['atomic_number','youngs_modulus']]
df3 = df2.sort_values(by=['youngs_modulus'],ignore_index=True)
display(df3)
```

Now divide the complete data into 7 subgroups of Youngs modulus. Suppress the groups 2, 4 and 6. Now you have only 4 groups of elements based on certain range of Youngs modulus. Now shuffle this data table
based on the Youngs modulus.

```python
import numpy as np
ymal = np.array_split(df3,7)
select = [ymal[0],ymal[2],ymal[4],ymal[6]]
df4 = pd.concat(select,ignore_index=True)
df5 = df4.sample(frac=1,ignore_index=True)
display(df5)
```

Apply k-Means method and show that there are 4 groups of elements

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
X = np.array(df5['youngs_modulus']).reshape(-1,1)
inertias = []
K = range(2, 10)
for k in K:
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    # Find the distance of points from centroid in each cluster and add all the distances
    inertias.append(k_means.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

Clearly, the elbow occurs in plot at K = 4, indicating 4 clusters

```python
k_means = KMeans(n_clusters=4)
k_means.fit(X)
centroids = k_means.cluster_centers_
labels = k_means.labels_
plt.plot(X[labels==0,0],len(X[labels==0,0])*[1],'r.', label='cluster 1')
plt.plot(X[labels==1,0],len(X[labels==1,0])*[1],'b.', label='cluster 2')
plt.plot(X[labels==2,0],len(X[labels==2,0])*[1],'g.', label='cluster 3')
plt.plot(X[labels==3,0],len(X[labels==3,0])*[1],'c.', label='cluster 4')
plt.plot(centroids[:,0],len(centroids[:,0])*[1],'mo',markersize=8, label='centroids')
plt.legend(loc='best')
plt.xlabel('Youngs Modulus')
plt.show()
```

By this method, although we do obtain 4 clusters, these aren't the actual groupings. This is because K-Means doesn't output equal size clusters, but rather more natural groupings based on similar nearby values.

```python
import cv2
import os
vectors = []
directory = '/content/drive/MyDrive/ME793/ME793_micrographs_small/'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    img = cv2.imread(f)
    vect = img.flatten()
    vectors.append(vect)
idf = pd.DataFrame(np.array(vectors))
display(idf) # dataframe with vectorised image data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_idf = scaler.fit_transform(idf)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.99)
tdf = pca.fit_transform(scaled_idf)
new_df = pd.DataFrame(tdf)
display(new_df) # feature reduction using PCA
```

```python
import warnings
warnings.filterwarnings('ignore')
inertias = []
K = range(2, 10)
for k in K:
    k_means = KMeans(n_clusters=k)
    k_means.fit(new_df)
    # Find the distance of points from centroid in each cluster and add all the distances
    inertias.append(k_means.inertia_)
```

```python
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

No clear elbow --> no obvious no of clusters

```python
images = []
directory = '/content/drive/MyDrive/ME793/ME793_micrographs_small/'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    img = cv2.imread(f)
    images.append(img)
all_imgs = np.array(images)
```

```python
all_imgs.shape
```

```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
model = VGG16(weights='imagenet',include_top=False,input_shape=(522,645,3))
features = model.predict(all_imgs)
features.shape # transfer learning for feature extraction
```

```python
vecs = []
for i in range(15):
  vec = features[i].flatten()
  vecs.append(vec)
ivecs = np.array(vecs)
ivecs.shape
```

```python
scaler = MinMaxScaler()
scaled_ivecs = scaler.fit_transform(ivecs)
```

```python
pca = PCA(n_components = 0.99)
ivdf = pca.fit_transform(scaled_ivecs)
new_ivdf = pd.DataFrame(ivdf)
display(new_ivdf)
```

```python
warnings.filterwarnings('ignore')
inertias = []
K = range(2, 10)
for k in K:
    k_means = KMeans(n_clusters=k)
    k_means.fit(new_ivdf)
    inertias.append(k_means.inertia_)
```

```python
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

No clear elbow --> no obvious no of clusters

```python
from sklearn.manifold import TSNE
model = TSNE(n_components = 2, random_state = 0, perplexity = 10)
tsne_data = model.fit_transform(scaled_ivecs)
tsne_data = np.vstack((tsne_data.T)).T
tsne_df = pd.DataFrame(data = tsne_data,
     columns =("Dim_1", "Dim_2"))
import seaborn as sn
sn.FacetGrid(tsne_df).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

No clustering visible in tSNE plot as well

```python
pca = PCA()
pca.fit(scaled_ivecs)
variances = pca.explained_variance_ratio_
import numpy as np
cum_vars = np.cumsum(variances)
plt.figure(figsize=(12,6))
plt.plot(range(1,16),cum_vars)
plt.xlabel('No of Components')
plt.ylabel('Variance Explained')
```

```python
# Let us reduce the PCA space to 3D (which is the max that can be visualised)
pca = PCA(3)
ivdf = pca.fit_transform(scaled_ivecs)
new_ivdf = pd.DataFrame(ivdf)
warnings.filterwarnings('ignore')
inertias = []
K = range(2, 10)
for k in K:
    k_means = KMeans(n_clusters=k)
    k_means.fit(new_ivdf)
    inertias.append(k_means.inertia_)
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

Elbow clearly occurs at K = 4, indicating 4 groups!

```python
k_means = KMeans(n_clusters=4)
k_means.fit(new_ivdf)
centroids = k_means.cluster_centers_
labels = k_means.labels_
X = new_ivdf
ax = plt.axes(projection ="3d")
ax.scatter3D(X.iloc[labels==0,0],X.iloc[labels==0,1],X.iloc[labels==0,2],'r.', label='cluster 1')
ax.scatter3D(X.iloc[labels==1,0],X.iloc[labels==1,1],X.iloc[labels==1,2],'b.', label='cluster 2')
ax.scatter3D(X.iloc[labels==2,0],X.iloc[labels==2,1],X.iloc[labels==2,2],'g.', label='cluster 3')
ax.scatter3D(X.iloc[labels==3,0],X.iloc[labels==3,1],X.iloc[labels==3,2],'c.', label='cluster 4')
ax.scatter3D(centroids[:,0], centroids[:,1], centroids[:,2], 'mo', label='centroids')
plt.legend(loc='best')
ax.set_xlabel('PCA_1')
ax.set_ylabel('PCA_2')
ax.set_zlabel('PCA_3')
ax.set_title('Clusters with centroids')
plt.show()
```

```python
# Let us further reduce the PCA space to 2D
pca = PCA(2)
ivdf = pca.fit_transform(scaled_ivecs)
new_ivdf = pd.DataFrame(ivdf)
warnings.filterwarnings('ignore')
inertias = []
K = range(2, 10)
for k in K:
    k_means = KMeans(n_clusters=k)
    k_means.fit(new_ivdf)
    inertias.append(k_means.inertia_)
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

Elbow clearly occurs at K = 3, indicating 3 groups!

```python
k_means = KMeans(n_clusters=3)
k_means.fit(new_ivdf)
centroids = k_means.cluster_centers_
labels = k_means.labels_
X = new_ivdf
plt.plot(X.iloc[labels==0,0],X.iloc[labels==0,1],'r.', label='cluster 1')
plt.plot(X.iloc[labels==1,0],X.iloc[labels==1,1],'b.', label='cluster 2')
plt.plot(X.iloc[labels==2,0],X.iloc[labels==2,1],'g.', label='cluster 3')
plt.plot(centroids[:,0],centroids[:,1],'mo',markersize=8, label='centroids')
plt.legend(loc='best')
plt.xlabel('PCA_1')
plt.ylabel('PCA_2')
plt.title('Clusters with centroids in PCA space')
plt.show()
```
