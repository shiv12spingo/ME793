```python
import scipy.io
from time import time
import scipy.linalg as la # numpy does not have LU decomposition
import matplotlib.pyplot as plt
import numpy as np
```

```python
sizes = []
inv_times = []
lu_times = []
for i in range(5):
    matrix = scipy.io.loadmat("C:/Users/Shivprasad/Documents/MATLAB/matrix_"+str(i+1)+".mat")['A']
    sizes.append(len(matrix))
    start = time()
    B = la.inv(matrix)
    end = time()
    inv_times.append(end-start)
    start = time()
    P,L,U = la.lu(matrix)
    end = time()
    lu_times.append(end-start)
times = scipy.io.loadmat("C:/Users/Shivprasad/Documents/MATLAB/var_times.mat")
inv_times_2 = times['inv_time']
lu_times_2 = times['lu_time']
times = [inv_times,lu_times,inv_times_2,lu_times_2]
times = [[x*1000000 for x in item] for item in times]
```

```python
plt.plot(sizes,np.squeeze(times[0]),label='Python')
plt.plot(sizes,np.squeeze(times[2]),label='MATLAB')
plt.xlabel('Matrix Size')
plt.ylabel('Time (microseconds)')
plt.title('Inverse Calculation')
plt.legend()
```

```python
plt.plot(sizes,np.squeeze(times[1]),label='Python')
plt.plot(sizes,np.squeeze(times[3]),label='MATLAB')
plt.xlabel('Matrix Size')
plt.ylabel('Time (microseconds)')
plt.title('LU Decomposition')
plt.legend()
```

Both python and matlab use LU decomposition to calculate inverse of matrices
