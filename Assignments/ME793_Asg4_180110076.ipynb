###c)

```python
def cal_cost(theta,X,y):
    '''
    Calculates the cost for given X and y
    theta = Vector of thetas 
    X     = Row of X's np.zeros((m,j))
    y     = Actual y's np.zeros((m,1))
    where:
        j is the no of features
    '''
    m = len(y)
    predictions = X.dot(theta)
    cost = np.sum(np.square(predictions-y))/(2*m)
    return cost
```

The gradient descent function is modified to include the additional dimension

```python
def gradient_descent(X,y,theta,learning_rate=0.01,iterations=100):
    '''
    X = Matrix of X with added bias units
    y = Vector of Y
    theta = Vector of thetas np.random.randn(j,1)
    iterations = no of iterations
    Returns the final theta vector and array of cost history over no of iterations
    '''
    m = len(y)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations,3))
    for it in range(iterations):
        prediction = np.dot(X,theta)
        theta = theta -(1/m)*learning_rate*(X.T.dot((prediction - y)))
        theta_history[it,:] = theta.T
        cost_history[it] = cal_cost(theta,X,y)
    return theta, cost_history, theta_history
```

The input data is scaled so that all features are equally treated for smooth optimisation

```python
n_iter = 1000
X_b = np.c_[np.ones((len(X),1)),X]
from sklearn.preprocessing import StandardScaler as SS
scaler = SS()
X_std = scaler.fit_transform(X_b)
```

```python
lr = 0.0001
theta = np.random.randn(3,1)
theta,cost_history,theta_history = gradient_descent(X_std,np.expand_dims(Y,1),theta,lr,n_iter)
print('For the standardised input, the predicted output is given by the linear model with\nIntercept = b0: {:0.3f},\nCoefficient = b1: {:0.3f},\nCoefficient = b2: {:0.3f}'.format(theta[0][0],theta[1][0],theta[2][0]))
print('Final cost (MSE): {:0.3f}'.format(cost_history[-1]))
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.set_ylabel('J(Theta)')
ax.set_xlabel('Iterations')
ax.plot(range(n_iter),cost_history,'b.')
```

It seems that learning rate = 0.01 showed the best training behaviour
