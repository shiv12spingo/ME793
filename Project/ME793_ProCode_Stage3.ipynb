###Feature Reduction

```python
#Removing columns that are highly correlated
corr_matrix = X3.corr()
drop_cols = []
for x in corr_matrix.index:
    for y in corr_matrix.index:
        if y == x:
            break
        corr_value = corr_matrix.loc[x, y]
        #Remove columns that have more than more than 90% correlation
        if abs(corr_value) >= 0.90:
            print(x, "|", y, "|", round(corr_value, 2)) 
            drop_cols.append(y)
```

```python
#Drop one of each pair of correlated columns
X4 = X3.drop(columns=drop_cols)
display(X4)
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X5 = pd.DataFrame(SS.fit_transform(X4))
display(X5) # standardized dataframe
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X5)
```

```python
variances = pca.explained_variance_ratio_
import numpy as np
cum_vars = np.cumsum(variances)
plt.figure(figsize=(12,6))
plt.plot(range(1,348),cum_vars)
plt.xlabel('No of Components')
plt.ylabel('Variance Explained')
```

The top 200 principal components explain almost all the variance

```python
pca_200 = PCA(n_components=200)
X6 = pd.DataFrame(pca_200.fit_transform(X5))
display(X6)
```

###Cross-Validated Hyperparameter Tuning

```python
lr_parameter_dict = {'solver': ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'], 'C' : [100,10,1,0.1,0.01]}
LRMO = lr()
from sklearn.model_selection import GridSearchCV
lr_gscv = GridSearchCV(estimator = LRMO, param_grid = lr_parameter_dict, n_jobs = -1, scoring = 'f1_macro', cv = 4, verbose=2)
lr_gscv.fit(X_train_2, Y_train_2)
lr_gscv.best_params_
```

```python
best_lrm = lr(C = 10, solver = 'newton-cg')
best_lrm.fit(X_train_2,Y_train_2)
Y_lr_best = best_lrm.predict(X_test)
print(classification_report(Y_test, Y_lr_best, target_names=['Good','Defective']))
```
